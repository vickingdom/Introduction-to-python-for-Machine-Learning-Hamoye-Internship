```python
#import libraries
import pandas as pd
import seaborn as sns
import numpy as np

#Read dataset
#energy_data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv')
energy_data = pd.read_csv('/home/victor/Desktop/energydata_complete.csv')


#print head()
print(energy_data.head())

#describe the dataset
print(energy_data.describe)

#check for null values
print(energy_data.isnull().sum())

#drop date and light column
energy_data = energy_data.drop(['date','lights'], axis = 1)

#print head
print(energy_data.head())

#Normalize the dataset using Min Max  scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

normalized_energy_data = pd.DataFrame(
    scaler.fit_transform(energy_data), columns = energy_data.columns)

#select Feature and Traget
independent_X = normalized_energy_data["T2"].values.reshape(-1,1)
dependent_Y = normalized_energy_data["T6"].values.reshape(-1,1)

#split the dataset into training and testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent_X, dependent_Y,
                                                    test_size = 0.30,
                                                    random_state = 42)

#import LinearRegression module
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()

#fit the training data
linear_model.fit(x_train, y_train)

#predictions
predicted_values = linear_model.predict(x_test)

#print prediction
print(predicted_values)
```

```python
# Q12
#import r^2 module
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values) 

#print r^2 score to 2 d.p
print("R^2 Value = ", round(r2_score, 2))    

```

```python
#Q13
#import MAE module
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)

#print mae to 2 d.p
print("MAE = ",(round(mae, 2)))
```

```python
#Q14
rss = np.sum(np.square(y_test - predicted_values))

#print rss to 2 d.p
print("RSS = ",round(rss, 2))

```

```python
#Q15
#import RMSE module
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))

#print rmse to 2 d.p
print("RMSE = ", round(rmse, 2))
      
```

```python
#Q17

#import libraries
import pandas as pd
import seaborn as sns
import numpy as np

#Read dataset
#energy_data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv')
energy_data = pd.read_csv('/home/victor/Desktop/energydata_complete.csv')

#drop date and light column
energy_data = energy_data.drop(['date','lights'], axis = 1)


#Normalize the dataset using Min Max  scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

normalized_energy_data = pd.DataFrame(
    scaler.fit_transform(energy_data), columns = energy_data.columns)

#select feature
#independent_X = normalized_energy_data.drop(columns =['Appliances'])

#dependent_Y = normalized_energy_data['Appliances']


X = normalized_energy_data.drop('Appliances',axis=1)
Y = normalized_energy_data['Appliances']


#split the dataset into training and testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y,
                                                    test_size = 0.30,
                                                    random_state = 42)

#import module
from sklearn.linear_model import LinearRegression
#LinearRegression 
linear_model = LinearRegression()
#Fit the training data
linear_model.fit(x_train, y_train)

#predictions
predicted_values = linear_model.predict(x_test)


#Get the weight of the linear model
def get_weights_df(model, feat, col_name) :
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = [ 'Features' , col_name]
    weights_df[col_name].round( 3 )
    return weights_df


linear_model_weights = get_weights_df(linear_model, x_train, 
                                               'Linear_Model_weights')
print(linear_model_weights)
# RH_2 and RH_1 have the Lowest and Highest weights
```

```python
#Q18

#Ridge module
from sklearn.linear_model import Ridge
ridge_regression = Ridge(alpha= 0.4 )
ridge_regression.fit(x_train, y_train)

#Train prediction
ridge_predictions_train = ridge_regression.predict(x_train)

#Test prediction
ridge_predictions_test = ridge_regression.predict(x_test)

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_train,ridge_predictions_train ))
#print rmse to 2 d.p
print("RMSE Train = ", round(rmse, 2))

rmse = np.sqrt(mean_squared_error(y_test, ridge_predictions_test))
#print rmse to 2 d.p
print("RMSE Test= ", round(rmse, 2))

#No
      
```

```python
#Q19
from sklearn.linear_model import Lasso
lasso_regression = Lasso(alpha= 0.001 )
lasso_regression.fit(x_train, y_train)

#Train prediction
lasso_predictions_train = lasso_regression.predict(x_train)

#Test prediction
lasso_predictions_test = lasso_regression.predict(x_test)


#Get the weight of the linear model
def get_weights_df(model, feat, col_name) :
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = [ 'Features' , col_name]
    weights_df[col_name].round( 3 )
    return weights_df

lasso_weights_energy_data = get_weights_df(lasso_regression, x_train, 'Lasso_weight')
print(lasso_weights_energy_data)

#4
```
